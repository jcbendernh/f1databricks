# Bronze to Silver - All Tables
This notebook transforms the .csv data in bronze volume to Delta tables in the silverdb 

## Circuits

```python
volume_file_path = "/Volumes/catadb360dev/default/f1/circuits.csv"

df_circuits = (
    spark.read
    .format("csv")
    .option("header", True)
    .option("inferSchema", True)
    .load(volume_file_path)
)

display(df_circuits)
```

```python
df_circuits.write.format("delta").mode("overwrite").saveAsTable("catadb360dev.f1silver.circuits")
```
